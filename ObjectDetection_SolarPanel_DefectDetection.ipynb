```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # or yolov8s.pt for a stronger model
model.train(data="/content/drive/MyDrive/Faulty_solar_panel/data.yaml", epochs=50, imgsz=640, batch=16)
```

```python
from ultralytics import YOLO

model = YOLO("/content/best.pt")
results = model(r"/content/drive/MyDrive/Faulty_solar_panel/testimg2.jpg", save=True)
results[0].show()
```
